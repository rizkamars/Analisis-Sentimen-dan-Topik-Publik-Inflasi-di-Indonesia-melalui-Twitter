# Preprocessing

```python
# Install library Sastrawi
!pip install Sastrawi

# Import library
import re
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import nltk

# Download stopwords NLTK
nltk.download('stopwords')

# Stopwords bahasa Indonesia
stop_words = set(nltk.corpus.stopwords.words('indonesian'))

# Tambahkan stopwords kustom
custom_stopwords = ['di', 'yg', 'aja', 'ya', 'ga','udah','udh','gak','gue','tp']
stop_words.update(custom_stopwords)

# Membaca file CSV dan mengambil 2000 baris pertama
df_pre = pd.read_csv('/content/ekonomi_filtered.csv').head(2000)

# Tampilkan beberapa baris awal data
df_pre.head()
```

```python
# Fungsi case folding
def case_folding(text):
    if isinstance(text, str):
        return text.lower()
    return ""

# Terapkan fungsi case folding
df_pre['case_folding'] = df_pre['full_text'].apply(case_folding)
df_pre.head()
```

```python
# Fungsi untuk menghapus tanda baca
def remove_punctuation(text):
    return re.sub(r'[^a-z\s]', '', text)

# Terapkan fungsi remove punctuation
df_pre['remove_punctuation'] = df_pre['case_folding'].apply(remove_punctuation)
df_pre.head()
```

```python
# Fungsi tokenizing
def tokenize(text):
    if isinstance(text, str):
        return text.split()
    return []

# Terapkan fungsi tokenizing
df_pre['tokenized'] = df_pre['remove_punctuation'].apply(tokenize)
df_pre.head()
```

```python
# Fungsi untuk menghapus stopwords
def remove_stopwords(tokens):
    if isinstance(tokens, list):
        return [word for word in tokens if word not in stop_words]
    return []

# Terapkan fungsi remove stopwords
df_pre['no_stopwords'] = df_pre['tokenized'].apply(remove_stopwords)
df_pre.head()
```

```python
# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming
def stemming(tokens):
    if isinstance(tokens, list):
        # Stemming pada setiap kata
        stemmed_words = [stemmer.stem(word) for word in tokens]
        return stemmed_words
    return []

# Terapkan fungsi stemming ke kolom 'no_stopwords'
df_pre['stemmed'] = df_pre['no_stopwords'].apply(stemming)

# Tampilkan beberapa baris setelah stemming diterapkan
df_pre.head()
```

```python
# Gabungkan kembali token setelah stemming menjadi satu string
df_pre['processed_full_text'] = df_pre['stemmed'].apply(lambda x: " ".join(x))

# Simpan hasil ke file baru
df_pre.to_csv('ekonomi_processed.csv', index=False)

print("Hasil preprocessing telah disimpan dalam 'ekonomi_processed.csv'")
```

# Analisis Deskriptif dan Visualisasi Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk


# Membaca file CSV yang sudah diproses
df = pd.read_csv('ekonomi_processed.csv')

# **1. Analisis Statistik Deskriptif**

# Jumlah berita
num_berita = df.shape[0]

# Menghitung jumlah kata pendek (panjang kata < 3)
df['word_count'] = df['processed_full_text'].apply(lambda x: len(x.split()))
short_words = df['processed_full_text'].apply(lambda x: sum(1 for word in x.split() if len(word) < 3)).sum()

# Frekuensi kata
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = vectorizer.fit_transform(df['processed_full_text'])
words_freq = np.asarray(tfidf_matrix.sum(axis=0)).flatten()
words = vectorizer.get_feature_names_out()

# Dataframe untuk frekuensi kata
word_freq_df = pd.DataFrame(list(zip(words, words_freq)), columns=["word", "frequency"])
word_freq_df = word_freq_df.sort_values(by="frequency", ascending=False)

# Tampilkan statistik dasar
print(f"Jumlah Berita: {num_berita}")
print(f"Jumlah Kata Pendek: {short_words}")
print(word_freq_df.head())

# **2. Visualisasi Bar Chart (Frekuensi Kata)**
plt.figure(figsize=(10,6))
top_words = word_freq_df.head(10)
plt.barh(top_words['word'], top_words['frequency'], color='skyblue')
plt.xlabel('Frekuensi')
plt.ylabel('Kata')
plt.title('10 Kata Teratas Berdasarkan Frekuensi')
plt.gca().invert_yaxis()  # Invert Y-axis to display the most frequent words at the top
plt.show()

# **3. WordCloud (Visualisasi Kata)**
word_freq_dict = dict(zip(word_freq_df['word'], word_freq_df['frequency']))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_dict)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis
plt.title('WordCloud - Kata-kata Teratas')
plt.show()
```

# Frekuensi TF-IDF

```python
# Import library yang diperlukan
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Membaca file CSV yang sudah diproses
df_processed = pd.read_csv('ekonomi_processed.csv')

# Pastikan kolom yang digunakan adalah 'processed_article_text'
processed_data = df_processed['processed_full_text']

# Inisialisasi TfidfVectorizer dengan max_features=1000
vectorizer = TfidfVectorizer(max_features=1000)

# Terapkan TF-IDF ke data yang sudah diproses
tfidf_matrix = vectorizer.fit_transform(processed_data)

# Ambil istilah/fitur TF-IDF
tfidf_terms = vectorizer.get_feature_names_out()

# Mengonversi matrix TF-IDF ke format Gensim corpus (list of tuples)
corpus = []
for doc in tfidf_matrix:
    corpus.append([(idx, doc[0, idx]) for idx in range(doc.shape[1]) if doc[0, idx] > 0])

# Simpan corpus dan terms untuk digunakan di bagian LDA
corpus_file = 'tfidf_corpus.mm'
dictionary_file = 'tfidf_dictionary.dict'

# Simpan corpus dan dictionary menggunakan Gensim
from gensim import corpora
corpora.MmCorpus.serialize(corpus_file, corpus)
dictionary = corpora.Dictionary([[term] for term in tfidf_terms])
dictionary.save(dictionary_file)

print("TF-IDF telah dihitung dan disimpan.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ambil skor TF-IDF dan kata-katanya
tfidf_scores = np.array(tfidf_matrix.sum(axis=0)).flatten()
tfidf_words = tfidf_terms

# Urutkan kata berdasarkan skor TF-IDF
sorted_indices = np.argsort(tfidf_scores)[::-1]
top_n = 20  # Jumlah kata teratas yang ingin ditampilkan

# Pilih top_n kata dan skor TF-IDF-nya
top_words = tfidf_words[sorted_indices][:top_n]
top_scores = tfidf_scores[sorted_indices][:top_n]

# Visualisasikan dengan bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_words, top_scores, color='skyblue')
plt.xlabel('Skor TF-IDF')
plt.title(f'Top {top_n} Kata dengan Skor TF-IDF Tertinggi')
plt.gca().invert_yaxis()  # Membalik urutan agar kata teratas berada di atas
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a dictionary of terms and their corresponding TF-IDF scores
tfidf_scores = {term: tfidf_matrix.getcol(idx).sum() for idx, term in enumerate(tfidf_terms)}

# Generate a word cloud based on TF-IDF scores
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(tfidf_scores)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from sklearn.decomposition import PCA

# Reduksi dimensi menggunakan PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(tfidf_matrix.toarray())

# Visualisasikan hasil PCA
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.6, c='blue', edgecolors='k')
plt.title('PCA dari TF-IDF Vektor')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

#  Latent Dirichlet Allocation (LDA)


```python
# Import library yang diperlukan
from gensim.models import LdaModel
from gensim import corpora

# Memuat corpus dan dictionary yang sudah disimpan
corpus = corpora.MmCorpus('tfidf_corpus.mm')
dictionary = corpora.Dictionary.load('tfidf_dictionary.dict')

# Model LDA
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)

# Menampilkan topik
for idx, topic in lda_model.print_topics(-1):
    print(f"Topik {idx}: {topic}")
```

```python
import matplotlib.pyplot as plt

# Menampilkan topik secara manual
num_topics = 5  # Jumlah topik yang digunakan

# Membuat subplots untuk setiap topik
fig, axes = plt.subplots(1, num_topics, figsize=(20, 6))

for idx in range(num_topics):
    # Ambil kata-kata dan bobotnya untuk setiap topik
    topic_words = lda_model.get_topic_terms(idx, topn=10)

    # Menyiapkan kata-kata dan bobot untuk ditampilkan
    words = [dictionary[word_id] for word_id, _ in topic_words]
    weights = [weight for _, weight in topic_words]

    # Visualisasi topik dengan bar plot
    axes[idx].barh(words, weights, color='skyblue')
    axes[idx].set_title(f'Topik {idx}')
    axes[idx].invert_yaxis()  # Agar kata teratas berada di atas
    axes[idx].set_xlabel('Bobot Kata')

# Tampilkan plot
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Jumlah topik yang digunakan
num_topics = 5

# Membuat subplots untuk setiap topik
fig, axes = plt.subplots(1, num_topics, figsize=(20, 6))

# Menampilkan WordCloud untuk setiap topik
for idx in range(num_topics):
    # Ambil kata-kata dan bobotnya untuk setiap topik
    topic_words = lda_model.get_topic_terms(idx, topn=20)

    # Membuat dictionary kata dan bobot untuk WordCloud
    word_freq = {dictionary[word_id]: weight for word_id, weight in topic_words}

    # Membuat WordCloud untuk topik ini
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

    # Menampilkan WordCloud
    axes[idx].imshow(wordcloud, interpolation='bilinear')
    axes[idx].axis('off')  # Matikan sumbu
    axes[idx].set_title(f'Topik {idx}')

# Tampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Memastikan teks sudah menjadi list of tokenized words
df['processed_full_text_tokenized'] = df['processed_full_text'].apply(lambda x: x.split())

# Fungsi untuk menghitung nilai koherensi
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=1):
    coherence_values = []
    for num_topics in range(start, limit, step):
        model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=10)
        coherence_model = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherence_model.get_coherence())
    return coherence_values

# Menyediakan corpus yang sudah diproses dan tokenized
texts = df['processed_full_text_tokenized'].tolist()

limit = 15
coherence_values = compute_coherence_values(dictionary, corpus, texts, limit)

# Plotting
plt.plot(range(2, limit), coherence_values)
plt.xlabel("Jumlah Topik")
plt.ylabel("Coherence Score")
plt.title("Evaluasi Jumlah Topik")
plt.show()
```
